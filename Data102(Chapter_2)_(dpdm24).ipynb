1. count the customer from each branch based on their type

````python
data_A = data[data['Branch'] == 'A']
data_A
````

````python
data_B = data[data['Branch'] == 'B']
data_B
````

````python
data_C = data[data['Branch'] == 'C']
data_C
````

````python
count_typeA = data_A.groupby('Customer type')['Invoice ID'].count().reset_index()
count_typeA
````

````python
count_typeB = data_B.groupby('Customer type')['Invoice ID'].count().reset_index()
count_typeB
````

````python
count_typeC = data_C.groupby('Customer type')['Invoice ID'].count().reset_index()
count_typeC
````

2. Make Customer count data frame

````python
d = {'Customer type' : count_typeA['Customer type'],'Total Customer A' : count_typeA['Invoice ID'],'Total Customer B' : count_typeB['Invoice ID'],'Total Customer C' : count_typeC['Invoice ID'] }
cust_count = pandas.DataFrame(data=d)
````

````python
cust_count
````

3. Make the grouped barchart

````python
import numpy as np
width = 0.2
np.array(range(len(cust_count)))-width
````

````python
width = 0.2

plt.bar(range(len(cust_count)), cust_count['Total Customer B'],width,color='#8FBC8F',label='Branch B')
plt.bar(np.array(range(len(cust_count)))-width, cust_count['Total Customer A'],width,color='#8B2323',label='Branch B')
plt.bar(np.array(range(len(cust_count)))+width, cust_count['Total Customer C'],width,color='#EEC900',label='Branch C')
plt.title('Customer type in different branch')
plt.ylabel('Total Customer')
plt.xlabel('Customer Type')
plt.xticks(range(len(cust_count)),cust_count['Customer type'])
plt.legend(loc='lower center');
````
